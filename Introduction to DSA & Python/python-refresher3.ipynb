### 6. List comprehension

```python
result= []

for i in range(1, 11):
    if i%2!=0:
        result.append(i)

result
```

```python
list(range(1,10,2))
```

```python
seq = range(1,11)

[a for a in seq if a % 2 != 0]
```

```python
ages = [23,34,45,56,67,78,89]
```

```python
for age in ages:
    if age > 30:
        print('age is greater than 30')
    else: 
        print('age is less than or equal to 30')
```

```python
['Greater than 30' if age > 30 else 'Less than or equal to 30' for age in ages]
```

### 7. Classes

```python
class MyRectangle:
    #Follow the camel-case for naming of the class
    def __init__(self, length, breadth): # method is a function define inside a class
        self.length = length
        self.breadth = breadth
        self.random_attribute = 0
        self.another_random_attribute= 'abc'
    
    def get_perimeter(self):
        return 2* (self.length + self.breadth)
    
    def get_area(self):
        return self.length * self.breadth
    
    def get_length(self):
        return self.length
    
    def get_breadth(self):
        return self.breadth
    
```

```python
obj1= MyRectangle(length=5, breadth=6)
```

```python
obj1.length
```

```python
obj1.get_area() # getting the area of the rectangle using the method defined in the class
```

```python
MyRectangle.get_area(obj1) # another way to call the method get_area of the class MyRectangle by passing the object as an argument
# this works behind the scenes when we call obj1.get_area() because when we call a method using the object, python automatically passes the object as the first argument to the method.
```

`self` is a convention in python to refer to the instance of the class. It is used to access the attributes and methods of the class. It is not a keyword in python, but it is a convention to use `self` as the first parameter of the methods defined in the class.
