## 1. Variables & Data Types

```python
# Python is dynamically typed — no need to declare types
x = 10          # int
y = 3.14        # float
name = "Alice"  # str
is_valid = True # bool

# Python integers have arbitrary precision — no overflow like C/Java
# This matters in DSA problems involving large factorials or Fibonacci numbers
big = 2 ** 1000  # works fine in Python, would overflow in most other languages

# None is Python's null — used as a default "empty" value (e.g. linked list nodes)
node_next = None

# type() lets you inspect types at runtime — useful for debugging
print(type(x), type(y), type(name))
```

## 2. range()

```python
# range() is memory efficient — it generates numbers on the go
# instead of storing the entire list in memory like [0,1,2,...,n]
# range(1000000) uses ~48 bytes; list(range(1000000)) uses ~8 MB
for i in range(5):         # 0 to 4 (stop is exclusive)
    print(i)

for i in range(2, 10, 2):  # start=2, stop=10, step=2
    print(i)               # prints 2 4 6 8

# Reverse iteration without creating a reversed list
for i in range(4, -1, -1): # counts down: 4 3 2 1 0
    print(i)
```

## 3. Lists

```python
# Lists are ordered, mutable, and allow duplicates
# Backed by a dynamic array — O(1) access by index, O(n) search
nums = [3, 1, 4, 1, 5, 9]

nums.append(2)    # O(1) amortized — adds to end
nums.insert(0, 0) # O(n) — shifts all elements right; avoid in tight loops
nums.pop()        # O(1) — removes last element
nums.pop(0)       # O(n) — removes first element, shifts everything left

print(nums[2])    # O(1) index access
print(nums[-1])   # negative index: last element (-2 = second last, etc.)

# Slicing creates a NEW list — O(k) where k = slice length
sub = nums[1:4]   # elements at index 1, 2, 3 (end index is exclusive)

# Lists can hold mixed types, but in DSA you usually keep one type for clarity
mixed = [1, "hello", 3.14, True]
```

## 4. Tuples

```python
# Tuples are immutable lists — cannot be changed after creation
# Faster than lists and can be used as dictionary keys (lists cannot)
point = (3, 7)
x, y = point  # unpacking — clean way to assign multiple variables

# Gotcha: a single-element tuple needs a trailing comma
not_a_tuple = (5)    # this is just the integer 5
is_a_tuple  = (5,)   # this is a tuple with one element

# Useful for returning multiple values from a function
def min_max(lst):
    return min(lst), max(lst)  # implicitly returns a tuple

lo, hi = min_max([4, 2, 9, 1])
print(lo, hi)  # 1 9

# Tuples as dict keys — useful for 2D grid coordinates in graph problems
visited = {}
visited[(0, 0)] = True  # mark cell (row=0, col=0) as visited
```

## 5. Dictionaries

```python
# Dictionary uses a hash table — O(1) average for get, set, delete
# Great for frequency counts, memoization, and graph adjacency maps in DSA
freq = {}
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]

for word in words:
    freq[word] = freq.get(word, 0) + 1  # .get(key, default) avoids KeyError

print(freq)  # {"apple": 3, "banana": 2, "cherry": 1}

# Python 3.7+ dicts maintain insertion order — useful when order matters

# Keys must be hashable (immutable) — strings, ints, tuples work; lists do NOT
# bad_key = {[1,2]: "value"}  # TypeError: unhashable type: list

# Check membership with "in" — O(1) vs O(n) for lists
if "apple" in freq:
    print("found")
```

## 6. Sets

```python
# Sets store unique elements using hashing — O(1) average for add/remove/lookup
# Use a set when you only care whether something exists, not how many times
seen = set()
nums = [1, 2, 3, 2, 1, 4]

for n in nums:
    if n not in seen:  # O(1) — much faster than "if n not in list"
        seen.add(n)

print(seen)  # {1, 2, 3, 4} — order is NOT guaranteed

# Convert a list to a set to remove duplicates instantly — O(n)
unique = set([1, 2, 2, 3, 3, 3])  # {1, 2, 3}

# Set operations useful in DSA (all O(n)):
a = {1, 2, 3}
b = {2, 3, 4}
print(a & b)  # intersection: {2, 3}
print(a | b)  # union: {1, 2, 3, 4}
print(a - b)  # difference: {1}
```

## 7. List Comprehensions & Generators

```python
# Concise way to build lists — same O(n) complexity as a for loop
squares = [x**2 for x in range(10)]              # O(n)
evens   = [x for x in range(20) if x % 2 == 0]  # with filter condition

# Generator expression — uses () instead of []
# Lazy: produces values one at a time, never stores the whole sequence in memory
gen = (x**2 for x in range(10))  # memory efficient for large sequences

# sum() can consume a generator directly — no intermediate list created
total = sum(x**2 for x in range(1000000))  # efficient — O(1) extra space

print(squares)
print(evens)
```

## 8. Functions

```python
# Functions are first-class objects — can be passed as arguments
def factorial(n):
    # Base case stops the recursion — without it, you get infinite recursion
    # and a RecursionError (Python default limit is ~1000 calls deep)
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)  # O(n) time, O(n) space on the call stack

print(factorial(5))  # 120

# Gotcha: mutable default arguments are shared across ALL calls
def bad_append(val, lst=[]):   # DO NOT do this — lst persists between calls
    lst.append(val)
    return lst

def good_append(val, lst=None):  # use None as sentinel instead
    if lst is None:
        lst = []
    lst.append(val)
    return lst
```

## 9. Sorting

```python
# Python uses Timsort — O(n log n) worst case, O(n) on nearly-sorted data
nums = [5, 2, 8, 1, 9]

nums.sort()                        # in-place, returns None — modifies original
sorted_nums = sorted(nums)         # returns a NEW list — original unchanged
desc = sorted(nums, reverse=True)  # descending order

# Sort by custom key — avoids writing a full comparator
words = ["banana", "kiwi", "apple", "fig"]
words.sort(key=len)  # sort by string length — key func called once per element
print(words)  # ["fig", "kiwi", "apple", "banana"]

# Sort tuples by second element — common in DSA (e.g. interval scheduling)
intervals = [(1, 4), (3, 5), (2, 3)]
intervals.sort(key=lambda x: x[1])  # sort by end time
print(intervals)  # [(2, 3), (1, 4), (3, 5)]
```

## 10. String Methods

```python
# Strings are immutable in Python — every modification creates a new string
s = "Hello, World!"

print(s.lower())               # lowercase — useful for case-insensitive comparisons
print(s.upper())               # uppercase
print(s.split(", "))           # split into list by delimiter
print(s.replace("World", "Python"))  # returns new string, does not modify s
print(s.strip())               # removes leading/trailing whitespace

# Use join() to build strings in a loop — NOT += 
# += creates a new string each time: O(n^2) total; join() is O(n)
chars = ["a", "b", "c"]
result = "".join(chars)        # efficient string building
print(result)                  # "abc"

# f-strings are the modern way to format strings (Python 3.6+)
name, score = "Alice", 95
print(f"{name} scored {score}/100")  # cleaner than .format() or %

# String slicing — same syntax as lists
print(s[0:5])    # "Hello"
print(s[::-1])   # reverse the string — common trick in DSA
```
