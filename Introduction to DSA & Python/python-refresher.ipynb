## 1. Variables & Data Types

```python
x = 10
y = 3.14
name = "Alice"
is_valid = True
node_next = None  # None is Python's null — used in linked list nodes

# Python integers never overflow (unlike C/Java) — safe for large factorials
big = 2 ** 1000
print(type(x), type(y), type(name))
```

## 2. range()

```python
# range() generates numbers on the go — uses ~48 bytes vs ~8 MB for list(range(1_000_000))
for i in range(5):          # 0 to 4 (stop is exclusive)
    print(i)

for i in range(2, 10, 2):   # start, stop, step
    print(i)

for i in range(4, -1, -1):  # count down without creating a reversed list
    print(i)
```

## 3. Lists

```python
nums = [3, 1, 4, 1, 5, 9]

nums.append(2)    # O(1) amortized
nums.insert(0, 0) # O(n) — shifts all elements right
nums.pop()        # O(1) — removes last
nums.pop(0)       # O(n) — removes first, shifts everything left

print(nums[2])    # O(1) index access
print(nums[-1])   # last element

sub = nums[1:4]   # slicing returns a NEW list — O(k) where k = slice length
```

## 4. Tuples

```python
# Immutable — faster than lists and can be used as dict keys
point = (3, 7)
x, y = point

not_a_tuple = (5)   # just the integer 5
is_a_tuple  = (5,)  # trailing comma makes it a tuple

# Tuple as dict key — handy for 2D grid coordinates in graph problems
visited = {(0, 0): True}

def min_max(lst):
    return min(lst), max(lst)

lo, hi = min_max([4, 2, 9, 1])
print(lo, hi)
```

## 5. Dictionaries

```python
# Hash table under the hood — O(1) average get/set/delete
freq = {}
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]

for word in words:
    freq[word] = freq.get(word, 0) + 1  # .get() avoids KeyError

print(freq)

# Keys must be immutable — ints, strings, tuples work; lists do NOT
# {[1, 2]: "val"}  --> TypeError: unhashable type: list

if "apple" in freq:  # O(1) lookup
    print("found")
```

## 6. Sets

```python
# O(1) average add/remove/lookup — prefer over lists for membership checks
seen = set()
nums = [1, 2, 3, 2, 1, 4]

for n in nums:
    if n not in seen:
        seen.add(n)

print(seen)  # order is NOT guaranteed

a, b = {1, 2, 3}, {2, 3, 4}
print(a & b)  # intersection: {2, 3}
print(a | b)  # union:        {1, 2, 3, 4}
print(a - b)  # difference:   {1}
```

## 7. List Comprehensions & Generators

```python
squares = [x**2 for x in range(10)]
evens   = [x for x in range(20) if x % 2 == 0]

# Generator expression: lazy — produces one value at a time, never stores all in memory
total = sum(x**2 for x in range(1_000_000))  # O(1) extra space

print(squares)
print(evens)
```

## 8. Functions & Recursion

```python
def factorial(n):
    if n <= 1:              # base case — without this you get infinite recursion
        return 1
    return n * factorial(n - 1)  # O(n) time and space (call stack)

print(factorial(5))  # 120

# Gotcha: mutable default arguments are shared across ALL calls
def good_append(val, lst=None):
    if lst is None:  # use None as sentinel instead of lst=[]
        lst = []
    lst.append(val)
    return lst
```

## 9. Sorting

```python
# Timsort — O(n log n) worst case, O(n) on nearly-sorted data
nums = [5, 2, 8, 1, 9]

nums.sort()                       # in-place, returns None
sorted_nums = sorted(nums)        # returns a new list, original unchanged
desc = sorted(nums, reverse=True)

words = ["banana", "kiwi", "apple", "fig"]
words.sort(key=len)
print(words)

# Sort by second element — common in interval scheduling problems
intervals = [(1, 4), (3, 5), (2, 3)]
intervals.sort(key=lambda x: x[1])
print(intervals)
```

## 10. Strings

```python
# Strings are immutable — every change creates a new object
s = "Hello, World!"
print(s.lower(), s.upper())
print(s.split(", "))
print(s.strip())         # remove leading/trailing whitespace

# join() over += in loops: += is O(n²) due to immutability; join() is O(n)
result = "".join(["a", "b", "c"])

name, score = "Alice", 95
print(f"{name} scored {score}/100")  # f-strings: cleaner than .format()

print(s[0:5])   # slicing
print(s[::-1])  # reverse a string — common DSA trick
```
