```python
def print_first_element(arr):
    print(arr[0])
```

[1.] Space complexity = O(n), Time complexity = O(1)

```python
def print_all_elements(arr):
    for element in arr:
        print(element)
```

[2.] Space complexity = O(n), Time complexity = O(n) 

```python
def print_all_pairs(arr):
    for i in arr:
        for j in arr:
            print(i, j)
```

[3.] Space complexity = O(n), Time complexity = O(n^2)

```python
def log_loop(n):
    while n > 1:
        print(n)
        n = n // 2
```

[4.] Space complexity = O(1), Time complexity= O(log n)

```python
def sort_list(arr):
    arr.sort()
```

[5.] Space complexity = O(n), Time Complexity = O(n ^ logn n)

```python
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)
```

[6.] Space complexity = O(1), Time Complexity = O(2^n)

```python
def create_zero_list(n):
    return [0]*n
```

[7.] Space complexity = O(1), Time Complexity = O(n)

```python
def sum_numbers(n):
    total = 0
    for i in range(n):
        total += i
    return total
```

[8.] Space complexity = O(1), Time Complexity = O(n)

```python
def print_element_twice(arr):
    for element in arr:
        print(element)
    for element in arr:
        print(element)
```

[9.] Space complexity = O(n), Time Complexity = O(n)

```python
def fixed_inner_loop(n):
    for i in range(n):
        for j in range(100):
            print(i+j)
```
