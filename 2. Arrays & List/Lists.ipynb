### Lists

##### Lists

```python
# creation

numbers = [10,20,30,40,50,60,70,80]
fruits = ['apple', 'banana', 'cherry', 'strawberry']
mixed_list = [10, 'apple', 3.14, True, None]
```

```python
type(numbers)
```

##### Access

```python
# access

numbers[2] # third value
```

##### Traversal

```python
# traversa;

for fruit in fruits:
    print(fruit, end =" ")
```

##### Insertion

```python
# insertion  

numbers.append(90) # add at the end
numbers
```

```python
numbers.insert(3, 35) # add at index 3
numbers
```

##### Merging

```python
# merging

li1 = [1,2,3,4,5]
li2 = [6,7,8]
```

```python
more_fruits = ['orange', 'grape', 'kiwi']
```

```python
fruits.extend(more_fruits) # merge more_fruits into fruits
fruits
```

##### Concatenation

```python
# concatenation

li1 + li2 # another way to merge lists
```

##### Deletion

```python
# deletion 

list = [1, 2, 3, 4, 5, 6, 22, 4, 6, 4]

list.remove(4) # remove first occurrence of 4
list
```

```python
list
```

```python
list.pop() # by default removes last element (first return and then remove)
list
```

```python
list.pop(2) # (first return and then remove)
```

```python
list
```

```python
del list[2] # delete element at index 2
list
```

```python
# full list -> empty list
list.clear()
list
```

```python
type(list) # it is still a list but empty
```

##### Updation

```python
# updation
 
mixed_list[1] = 'banana' # update index 1
mixed_list
```

##### Count

```python
li = [2,3,4,5,2,4,22,2,5,2,4,5,2,4,5,2,4,5]
li.count(2) # count number of occurrences of 2
```

##### Sort

```python
li.sort() # sort the list in ascending order
li
```

```python
li1 = [2,4,3,1,-4,5]
sorted(li1) # another way to sort a list (does not change original list), need to assign it to a variable if we want to store the sorted list
```

```python
li2= [3,4,1,2,-5,7,4]
li2.reverse() # reverse the list
li2
```

```python
li2 = li1.copy() # create a copy of li1 because lists are mutable and any chnages to it will affect it. So, by making a copy we insure that any change to it won't affect the original list.
li2 
```
