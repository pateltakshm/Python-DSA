### Practice coding problems on Arrays

#### 1. Find the largest number in a list

```python
numbers= [12,45,7,89,34]

max(numbers) # find the maximum number in the list
```

```python
# another way to find the maximum number in the list without using built-in functions

largest = numbers[0] # Initialize largest to the first element
for element in numbers[1:]:
    if element > largest:
        largest = element
print(f'largest element is {largest}') 

# time complexity of the above code is O(n) because we have to iterate through the entire list once to find the largest element.
```

#### 2. Count Occurence of an element

```python
items = ["apple", "bannana", "apple", "grape", "avocado", "apple", "strawberry", "blueberry", "raspberry"]
target = "apple"

items.count(target) # count the number of times the target appears in the list
```

```python
# now idea is to work without using built-in functions

count = 0
for item in items:
    if item == target:
        count += 1
print(f'{target} appears {count} times in the list')

# time complexity here too is O(n) because we have to iterate through the entire list once to count the occurrences of the target.
```

#### 3. Remove duplicate form the list

```python
numbers= [1, 2, 2, 3, 4, 5, 5, 5, 6, 7, 8, 9, 9, 10]

list(set(numbers)) # find the unique elements in the list
```

```python
list  = []

for n in numbers:
    if n not in list:
        list.append(n)
print(list)

# time complexity of the above code is O(n^2) because for each element in the numbers list, we are checking if it is already in the list of unique elements, which takes O(n) time. Therefore, the overall time complexity is O(n) * O(n) = O(n^2).
```
