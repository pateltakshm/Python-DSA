`Array` - An array is a data structure that stores a collection of elements of the same data type in contiguous memory locations.

### **1. Arrays**

```python
import array
```

```python
# creation

arr1 = array.array('i', [10, 20, 30, 40, 50])
print(arr1)
```

```python
# access

arr1[2]
```

```python
arr1[-2]
```

```python
# traversal
for element in arr1:
    print(element, end=' ')
```

```python
# insertion
arr1.insert(3, 35) # (index, value)
print(arr1)
```

```python
# deletion
arr1.remove(30) # remove the first occurrence of value
print(arr1)
```

```python
# updation
arr1[2] = 30 
print(arr1)
```

```python
arr2 = array.array('i', [60, 70, 80])
```

```python
# sorting
arr3= array.array('i', [4,6,3,9,5,2])
print(arr3)
```

```python
array.array('i', sorted(arr3)) # sorted one
```

```python
# merging
arr4 = arr1 + arr2 
print(arr4)
```

```python
# spliting
ix = int(len(arr4) / 2)
ix
```

```python
arr4[: ix] 
```

```python
arr4[ix :]
```

### **2. Numpy**

```python
import numpy as np
```

```python
# creation
np1= np.array([10, 20, 30, 40, 50])
np2= np.array([50, 60, 70, 80])
print(np1)
print(np2)
```

```python
# access
np1[2].item()
```

```python
# traversal
for element in np1:
    print(element, end=' ')
```

```python
# insertion
np1 = np.insert(np1, 3, 35) # (array, index, value)
print(np1)
```

```python
np.insert(np1, 3, 37)
```

```python
# deletion
np.delete(np1, 3) # (array, index)
```

```python
# updation
np1[2] = 30
print(np1)
```

```python
# sorting
np3 = np.array([4,6,3,9,5,2])
print(np3)
```

```python
np.sort(np3)
```

```python
# merging
np4= np.concatenate((np1, np2))
np4
```

```python
# splitting

np.array_split(np4, 2) 
```
