### 2D Operation

##### Creation

```python
# creation

matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

type(matrix) # it is still a list, just a list of lists
```

```python
def display_matrix(matrix): # display function
    for row in matrix:
        print(row)  
```

```python
display_matrix(matrix)
```

##### Traversal

```python
# traversal

for row in matrix:
    for element in row:
        print(element, end= ' ')
    print() # for new line after each row 
```

```python
# a safer way to traverse the matrix using indices

for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        print(matrix[i][j], end= ' ')
    print() # for new line after each row
```

```python
matrix[2][0] # access element at third row and first column
```

##### Updation

```python
# updation

matrix[0][1] = 20 # update element at first row and second column to 20
display_matrix(matrix)
```

```python
# insertion

matrix[2].insert(1, 50) # insert 50 at second column of third row
display_matrix(matrix)
```

```python
matrix[0].append(30) # append 30 at the end of first row
display_matrix(matrix)
```

##### Deletion

```python
# deletion

matrix[2].remove(50) # remove 50 from third row
display_matrix(matrix)
```

```python
matrix[0].pop() # remove last element from first row
display_matrix(matrix)
```

### 3D Operation

```python
threeD_list = [ [[1, 2, 3]], [[4, 5, 6]], [[7, 8, 9]] ]
type(threeD_list) # it is still a list, just a list of lists of lists
```

```python
for i in range(len(threeD_list)):
    for j in range(len(threeD_list[i])):
        for k in range(len(threeD_list[i][j])):
            print(threeD_list[i][j][k], end= ' ')
    print() # for new line after each row
```
