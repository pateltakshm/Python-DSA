### String Manipulation Techniques

#### 1. Case conversion

```python
str1= "this is DATA SCIENCE"
```

##### upper 

```python
str1.upper()
```

##### lower

```python
str1.lower()
```

##### Title

```python
str1.title() # capitalize first character of each words.
```

```python
name = "taksh patel"
name.title()
```

##### Capitalize

```python
name.capitalize() # capitalize only first letter of the string
```

##### Swapcase

```python
str1.swapcase() # turns lowercase to upper and vice versa
```

#### 2. White space handling

```python
str1= '     this is DATA SCIENCE   '
str1
```

##### strip

```python
str1.strip()
```

##### lstrip

```python
str1.lstrip() # only cleaned on the left side in terms of the white space
```

##### rstrip

```python
str1.rstrip() 
```

```python
str2 = '##@@$$$this is DATA SCIENCE####$$$$@@'
str2.rstrip("#$@")
```

```python
str1.find('Data') # it gives -1 if not found
```

##### index

```python
str1.index('data')
```

```python
str1.index('ML')
```

##### startswith

```python
str1.startswith('data')
```

```python
str1.startswith('this')
```

##### endswith

```python
str1.endswith('this')
```

```python
str1.endswith('science')
```
