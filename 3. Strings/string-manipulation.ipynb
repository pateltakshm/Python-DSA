#### String Comparison

##### Lexicographically

```python
str1= 'banana'
str2= 'banned'
```

```python
ord('a'), ord('n')
```

```python
str2 > str1 # 4th character is bigger in str2
```

```python
str3 = 'ball'
str4 = 'balls'
```

```python
str4 > str3
```

##### Case sensitivity

```python
str1= 'zoo'
str2= 'zoO'

str1 == str2
```

```python
for ch1, ch2 in zip(str1, str2):
    print(ch1, ord(ch1), '<->', ch2, ord(ch2))
```

##### sorting

```python
str1 = 'abcABC'
str1
```

```python
sorted(str1) # capital value comes before than small ones in ASCII
```

```python
for ch in str1:
    print(ch, '->', ord(ch))
```
