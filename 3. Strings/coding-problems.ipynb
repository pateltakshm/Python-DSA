#### 1. count vowels and consonents

```python
s= '__hel lo 123$world '

vowels= 0
consonents= 0

for char in s:
    if char.isalpha():
        if char in 'aeiou':
            vowels+=1
        else:
            consonents+=1 
print(f"There are {vowels} vowels and {consonents} consonents")
```

#### 3. Palindrome String

```python
s= 'madam'

# left to right is same. eg, 'malayalam'

s_rev = s[::-1]

if s == s_rev:
    print('Palindrome')
else:
    print('Not Palindrome')
```

#### 4. Check for Anagrams

* Return `True` if two strings contain the same characters in any order
* Strings will be of same length and lowercase

```python
s1= 'listen'
s2= 'silent'
```

##### Time: `nlog(n)`

```python
s1_sorted = sorted(s1)
s1_sorted
```

```python
s2_sorted= sorted(s2)
s2_sorted
```

```python
if s1_sorted == s2_sorted:
    print('Anagram')
else: 
    print('Not Anagram')
```

```python
mapping = [0] *26 
len(mapping)
```

```python
for ch in s1:
    index = ord(ch) - 97
    # print(ch, ord(ch), index)
    mapping[index] +=1

mapping
```

```python
for ch in s2:
    index = ord(ch) - 97
    # print(ch, ord(ch), index)
    mapping[index] -= 1
    
mapping
```

```python
if all(m == 0 for m in mapping):
    print('anagram')
else:
    print('Not anagram')
```
