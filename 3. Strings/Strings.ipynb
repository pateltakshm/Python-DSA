`Core Definitions`

`Mutable`: An object is mutable if you can change its internal data or properties without creating an entirely new object. It is modified "in place" at the same memory address.

`Immutable`: An object is immutable if its state cannot be modified once it is established. If you want to "change" an immutable object, the system must create a brand-new object with a new memory address to hold the updated data.
