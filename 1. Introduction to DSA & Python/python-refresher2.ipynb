### 3. Loops

#### 3.1 For Loop

```python
# if we know the no. of iterations, we can use for loop
list = [1,2,4,6,8]

print (list[0])
print (list[1])
print (list[2])
print (list[3])
print (list[4])
```

```python
for i in list: # independent of list size
    print(i)
```

```python
for i in range(20,31):
    if i % 2 == 0:
        print(i)
```

```python
for i in range(20,31,2): # we can use steps for even numbers
    print(i) 
```

#### 3.2 While loop

```python
# For condition based loops, we can use while loop
a= 1

while a <= 5:
    print(a)
    a+=1
```

### 5. Build in function

#### 5.1 Enumerate

```python
# i am not going to revise min, max, sum, len functions, because they are very basic.
names = ['Bruce', 'Peter', 'Isha', 'Taksh', 'Stuart']
```

```python
for i in range(len(names)):
    print(i, names[i]) 
```

```python
for x, name in enumerate(names, start=1): # enumerate is a built-in function that returns an enumerate object, which is an iterator that produces pairs of index and value from the input iterable. The start parameter is used to specify the starting index for enumeration. By default, it starts from 0, but in this case, we are starting from 1.
    print(x, name) 
```

#### 5.2 Zip

```python
colors = ['Red', 'Green', 'Blue', 'Yellow', 'Purple']
ages = [25, 30, 35, 40, 45]

for a, b, c in zip(names, colors, ages): # zip is a built-in function that takes two or more iterables and returns an iterator of tuples, where the i-th tuple contains the i-th element from each of the input iterables. The iterator stops when the shortest input iterable is exhausted.
    print(f'{a} is {c} years old and likes {b} color')
```
