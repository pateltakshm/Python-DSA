#### 2.2 if-elif-else

```python
marks = 45

if marks > 90:
    print("Excellent")
elif marks > 60 and marks <= 90:
    print("Good")
elif marks >50 and marks <= 60:
    print("Average")
else:
    print("Needs Improvement")
```
