## What is a Linked List?

A **Linked List** is a linear data structure where elements (called **nodes**) are stored in memory non-contiguously. Each node holds:

1. **Data** — the actual value stored.
2. **Pointer (next)** — a reference to the next node in the sequence.

Unlike Python's built-in `list` (which is a dynamic array), a linked list does **not** require contiguous memory. Nodes can live anywhere in memory and are connected via pointers.

```
[Head] → [Node 1 | next] → [Node 2 | next] → [Node 3 | None]
```

---

## Types of Linked Lists

| Type | Description |
|------|-------------|
| **Singly Linked List** | Each node points to the next node only. |
| **Doubly Linked List** | Each node points to both next and previous nodes. |
| **Circular Linked List** | The last node points back to the head. |

---

## Time & Space Complexity

| Operation | Linked List | Python List (Array) |
|-----------|-------------|---------------------|
| Access by index | O(n) | O(1) |
| Search | O(n) | O(n) |
| Insert at head | O(1) | O(n) |
| Insert at tail | O(n) | O(1) amortized |
| Insert at middle | O(n) | O(n) |
| Delete at head | O(1) | O(n) |
| Space | O(n) | O(n) |

## 1. Node Class

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None  # pointer to next node

# Create individual nodes
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)

# Link them manually
n1.next = n2
n2.next = n3

# Traverse
current = n1
while current:
    print(current.data, end=' -> ')
    current = current.next
print('None')
````

## 2. Singly Linked List — Full Implementation

````python
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete(self, data):
        if not self.head:
            return
        if self.head.data == data:
            self.head = self.head.next
            return
        current = self.head
        while current.next:
            if current.next.data == data:
                current.next = current.next.next
                return
            current = current.next

    def search(self, data):
        current = self.head
        index = 0
        while current:
            if current.data == data:
                return index
            current = current.next
            index += 1
        return -1

    def display(self):
        elements = []
        current = self.head
        while current:
            elements.append(str(current.data))
            current = current.next
        print(' -> '.join(elements) + ' -> None')

    def length(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count


# Demo
ll = SinglyLinkedList()
ll.append(10)
ll.append(20)
ll.append(30)
ll.prepend(5)
ll.display()

print('Search 20:', ll.search(20))
ll.delete(20)
ll.display()
print('Length:', ll.length())
````

## 3. Doubly Linked List

````python
class DNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None  # backward pointer


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = DNode(data)
        if not self.head:
            self.head = self.tail = new_node
            return
        new_node.prev = self.tail
        self.tail.next = new_node
        self.tail = new_node

    def display_forward(self):
        elements = []
        current = self.head
        while current:
            elements.append(str(current.data))
            current = current.next
        print('Forward: None <-> ' + ' <-> '.join(elements) + ' <-> None')

    def display_backward(self):
        elements = []
        current = self.tail
        while current:
            elements.append(str(current.data))
            current = current.prev
        print('Backward: None <-> ' + ' <-> '.join(elements) + ' <-> None')


dll = DoublyLinkedList()
for val in [1, 2, 3, 4, 5]:
    dll.append(val)

dll.display_forward()
dll.display_backward()
````

## 4. Reverse a Linked List (Classic Interview Problem)

````python
def reverse_linked_list(head):
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev  # new head


ll2 = SinglyLinkedList()
for val in [1, 2, 3, 4, 5]:
    ll2.append(val)

print('Before reversal:')
ll2.display()

ll2.head = reverse_linked_list(ll2.head)
print('After reversal:')
ll2.display()
````

## 5. Detect a Cycle — Floyd's Tortoise & Hare

````python
def has_cycle(head):
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False


# Build a cycle: 1 -> 2 -> 3 -> 4 -> 2 (cycle back to node with data=2)
a = Node(1); b = Node(2); c = Node(3); d = Node(4)
a.next = b; b.next = c; c.next = d; d.next = b  # cycle!

print('Has cycle:', has_cycle(a))  # True

# No cycle
x = Node(1); y = Node(2); z = Node(3)
x.next = y; y.next = z
print('Has cycle:', has_cycle(x))  # False
````

## 6. How Linked Lists Help in AI / ML

Linked lists are a foundational concept that enables many structures used in AI and ML:

---

### 6.1 Dynamic Computation Graphs (PyTorch Autograd)

PyTorch builds a **dynamic computation graph** during the forward pass — essentially a linked structure of operations. During `.backward()`, it traverses this graph (like traversing a linked list) to compute gradients.

```
Input → Linear → ReLU → Linear → Loss
  ↑        ↑       ↑       ↑       ↑
  node → node → node → node → node  (linked during forward pass)
```

---

### 6.2 LRU Cache (used in ML inference pipelines)

LRU (Least Recently Used) cache is implemented using a **Doubly Linked List + HashMap**. It's used in:
- Model serving to cache frequently requested predictions.
- KV-cache in LLM inference (e.g., caching attention keys/values for fast generation).

````python
from collections import OrderedDict

class LRUCache:
    """LRU Cache backed by an OrderedDict (doubly linked list internally)."""

    def __init__(self, capacity: int):
        self.cache = OrderedDict()
        self.capacity = capacity

    def get(self, key):
        if key not in self.cache:
            return -1
        self.cache.move_to_end(key)  # mark as recently used
        return self.cache[key]

    def put(self, key, value):
        if key in self.cache:
            self.cache.move_to_end(key)
        self.cache[key] = value
        if len(self.cache) > self.capacity:
            self.cache.popitem(last=False)  # evict least recently used


# Simulating an ML inference cache (e.g., caching predictions for user IDs)
inference_cache = LRUCache(capacity=3)
inference_cache.put('user_101', 0.92)  # predicted churn probability
inference_cache.put('user_202', 0.45)
inference_cache.put('user_303', 0.78)

print('Cached prediction for user_101:', inference_cache.get('user_101'))

inference_cache.put('user_404', 0.61)  # evicts user_202 (LRU)
print('user_202 after eviction:', inference_cache.get('user_202'))  # -1 (evicted)
````

### 6.3 Sequential Data Processing (NLP pipelines)

Before transformers, **Recurrent Neural Networks (RNNs)** processed tokens one by one — essentially walking through a linked sequence of hidden states:

```
h0 → h1 → h2 → h3 → ... → hn
 ↑    ↑    ↑    ↑           ↑
w1   w2   w3   w4          wn   (input tokens)
```

Each hidden state `h_t` depends on `h_{t-1}` — the same pointer-following pattern as a linked list.

---

### 6.4 Decision Trees and Graph Traversal

Decision trees are built from nodes that link to children — structurally similar to a linked list. Tree/graph traversal algorithms (DFS, BFS) used in:
- **Beam search** (decoding in LLMs)
- **Monte Carlo Tree Search** (used in AlphaGo, AlphaZero)

all rely on the same pointer-following logic as linked list traversal.

````python
# Simple Decision Tree Node (structurally a linked / tree node)
class TreeNode:
    def __init__(self, feature=None, threshold=None, value=None):
        self.feature = feature
        self.threshold = threshold
        self.value = value      # leaf prediction
        self.left = None
        self.right = None

    def predict(self, x):
        if self.value is not None:  # leaf node
            return self.value
        if x[self.feature] <= self.threshold:
            return self.left.predict(x)
        return self.right.predict(x)


# Manually build: if feature[0] <= 5 → class 0, else → class 1
root = TreeNode(feature=0, threshold=5)
root.left = TreeNode(value=0)
root.right = TreeNode(value=1)

samples = [{'feat': [3]}, {'feat': [7]}, {'feat': [5]}]
for s in samples:
    print(f"Input feature={s['feat'][0]} → Predicted class: {root.predict(s['feat'])}")
````

### 6.5 Memory Networks & Attention (Pointer Networks)

**Pointer Networks** (used in tasks like sorting, convex hull) use attention over a linked sequence of inputs to select the next output — literally "pointing" to nodes in a sequence, inspired by linked-list pointer logic.

## 7. Python's Built-in `collections.deque` — A Linked List Under the Hood

````python
from collections import deque

# deque is implemented as a doubly linked list of fixed-size blocks
# O(1) append and pop from both ends — used heavily in ML pipelines

# Example: sliding window for time-series feature extraction
window_size = 3
stream = [1.2, 3.4, 2.1, 5.5, 4.0, 6.7, 3.3]
window = deque(maxlen=window_size)

print('Sliding window averages:')
for val in stream:
    window.append(val)
    if len(window) == window_size:
        avg = sum(window) / window_size
        print(f'  window={list(window)}  avg={avg:.2f}')
````
